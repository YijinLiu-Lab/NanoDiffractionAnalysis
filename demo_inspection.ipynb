```python
import os
import time
import numpy as np
import torch
import itertools
import argparse
from glob import glob

from model import ReconstructionNet
from dataset import Dataset
from scipy.io import savemat
```

```python
dataset_root = '/content/drive/data'
dataset_name = 'diffraction'
num_points = 2048
split = 'test'
dataset = Dataset(root=dataset_root, dataset_name=dataset_name, 
                        num_points=num_points, split=split, load_name=False)
```

```python
def load_pretrain(model, pretrain):
    state_dict = torch.load(pretrain, map_location='cpu')
    from collections import OrderedDict
    new_state_dict = OrderedDict()
    for key, val in state_dict.items():
        if key[:6] == 'module':
            name = key[7:]  # remove 'module.'
        else:
            name = key
        new_state_dict[name] = val
    model.load_state_dict(new_state_dict)
    print(f"Load model from {pretrain}")
    return model   
```

```python
class Par:
    pass
args = Par()
args.encoder = 'foldnet'
args.k = 32
args.split = 'test'
args.task = 'reconstruct'
args.feat_dims = 512
args.shape='gaussian'

for key in range(int(len(dataset))):
  print(key)
  # key = 1500
  pts,lb = dataset[key]
  mdic = {"rawpts":pts.numpy()}
  savemat("/results_heat/raw_"+str(key)+".mat",mdic)

  model = ReconstructionNet(args)

  model_path = '/snapshot/Reconstruct_diffractiontest_20210509_foldnet/models/diffractiontest_best.pkl'
  model = load_pretrain(model, model_path)
  model.eval()
  reconstructed_pl, feature = model(pts.view(1, 2048, 3))
  mdic = {"recpts":reconstructed_pl[0].detach().numpy(),"feature":feature.detach().numpy()}
  savemat("/results_heat/rec_"+str(key)+".mat",mdic)
```
